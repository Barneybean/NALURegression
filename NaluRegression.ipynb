```python
import numpy as np 
import pandas as pd 

train = pd.read_csv('data/train.csv')
```

```python
train
```

```python
train.values.shape
```

```python
train_w_dummies = pd.get_dummies(train)
```

```python
train_w_dummies
```

```python
# Check number of na's in each column
for col in train_w_dummies.columns:
    print(col, np.sum(train_w_dummies[col].isna()))
```

```python
# na's only in LotFrontage so remove them, and id as its not needed
processed_train = train_w_dummies.drop(columns = ['Id', 'LotFrontage', 'MasVnrArea', 'GarageYrBlt'])
```

```python
processed_train
```

```python
# Get shape of data, which is 1460 training examples by 288 variables
processed_train.values.shape
```

```python
# create train test split 
from sklearn.model_selection import train_test_split

X = processed_train.drop(columns = 'SalePrice').values
y = processed_train['SalePrice'].values
```

```python
print(X.shape)
print(y.shape)
```

```python
# Create train test splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.1, random_state=42)
```

```python
# Divide targets by 100 to make it easier 
y_train = y_train / 1000
y_test = y_test / 1000
```

```python
# Create normal DNN model, Use relu activations as they are unbounded continuous functions 
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization
from keras.optimizers import RMSprop

rmsprop = RMSprop(lr = 0.01, decay = 0.001)

dnn1_model = Sequential()
dnn1_model.add(BatchNormalization(input_shape = [X_train.shape[1]]))
dnn1_model.add(Dense(1, activation = 'relu'))

dnn1_model.compile(loss = ['mean_squared_error'], optimizer = rmsprop)

dnn2_model = Sequential()
dnn2_model.add(Dense(32, input_shape = [X_train.shape[1]], activation = 'relu'))
dnn2_model.add(BatchNormalization())
dnn2_model.add(Dense(1, activation = 'relu'))

dnn2_model.compile(loss = ['mean_squared_error'], optimizer = rmsprop)
```

```python
print(dnn1_model.summary())
print(dnn2_model.summary())
```

```python
# train 1 layer dnn over 100 epochs 

epochs = 100

dnn1_train_acc = []
dnn1_test_acc = []

for epoch in range(epochs):
    print("Epoch:", epoch)
    permutation = np.random.permutation(len(y_train))
    dnn1_model.fit(X_train[permutation], y_train[permutation])
    dnn1_train_acc.append(dnn1_model.evaluate(X_train, y_train))
    dnn1_test_acc.append(dnn1_model.evaluate(X_test, y_test))
    print("Train acc -", dnn1_train_acc[-1])
    print("Test acc -", dnn1_test_acc[-1])
    print("")
```

```python
# Train two layer 
epochs = 100
batch_size = len(X_train)

dnn2_train_acc = []
dnn2_test_acc = []

for epoch in range(epochs):
    print("Epoch:", epoch)
    permutation = np.random.permutation(len(y_train))
    dnn2_model.fit(X_train[permutation], y_train[permutation])
    dnn2_train_acc.append(dnn2_model.evaluate(X_train, y_train))
    dnn2_test_acc.append(dnn2_model.evaluate(X_test, y_test))
    print("Train acc -", dnn2_train_acc[-1])
    print("Test acc -", dnn2_test_acc[-1])
    print("")
```

```python
# create Nalu model 
from nalu import NALU
from keras.layers import ThresholdedReLU

rmsprop_nalu = RMSprop(lr =0.01, decay = 1e-4, clipnorm = 0.3)

nalu_model = Sequential()
nalu_model.add(BatchNormalization(input_shape = [X_train.shape[1]]))
nalu_model.add(ThresholdedReLU())
nalu_model.add(NALU(1))

nalu_model.compile(loss = ['mean_squared_error'], optimizer = rmsprop_nalu)
```

```python
print(nalu_model.summary())
```

```python
# Train NALU model 
epochs = 100

nalu_train_acc = []
nalu_test_acc = []

for epoch in range(epochs):
    print("Epoch:", epoch)
    permutation = np.random.permutation(len(y_train))
    nalu_model.fit(X_train[permutation], y_train[permutation])
    nalu_train_acc.append(nalu_model.evaluate(X_train, y_train))
    nalu_test_acc.append(nalu_model.evaluate(X_test, y_test))
    print("Train acc -", nalu_train_acc[-1])
    print("Test acc -", nalu_test_acc[-1])
    print("")
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error as mse

lin_reg = LinearRegression()
lin_reg_train_acc = []
lin_reg_test_acc = []

lin_reg.fit(X_train, y_train)
lin_reg_train_acc.append(mse(lin_reg.predict(X_train), y_train))
lin_reg_test_acc.append(mse(lin_reg.predict(X_test), y_test))

lin_reg_train_acc = lin_reg_train_acc * epochs
lin_reg_test_acc = lin_reg_test_acc * epochs
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 100)

rf_train_acc = []
rf_test_acc = []

rf.fit(X_train, y_train)
rf_train_acc.append(mse(rf.predict(X_train), y_train))
rf_test_acc.append(mse(rf.predict(X_test), y_test))

rf_train_acc = rf_train_acc * epochs
rf_test_acc = rf_test_acc * epochs
```
